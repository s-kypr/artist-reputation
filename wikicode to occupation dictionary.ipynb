## Merge the dictionaries to get the code

Code to create dictionary: Q12334 -> "Artist"

```python
import json
```

```python
# used to read gender and occupations dictionaries
def read_dictionary(path):
    with open(path) as json_file:
        line = json_file.readline()
        dictionary = json.loads(line)
    return dictionary
    
# read occupations dictionary
def read_dict_occupations(path):
    dict_occupations = {}
    with open(path) as json_file:
        content = json_file.readlines()
        for line in content:
            occ = json.loads(line)
            dict_occupations.update(occ)
    return dict_occupations
```

```python
## Merge the 2 dictionaries to create one for code of occupation -> 
LOCAL_PATH = "./data/"
QCODE_OCCUPATIONS_DICT = LOCAL_PATH + "dict_occupations.json"
OCCUPATIONS_FIELD_DICT = LOCAL_PATH + "dict_categories_occupations.json"


occupation_to_field = read_dictionary(OCCUPATIONS_FIELD_DICT)       # singer -> Artist
code_to_occupation = read_dict_occupations(QCODE_OCCUPATIONS_DICT)  # Q123 -> singer
```

```python
# create dictionary with code to field: Q123 -> Artist
code_to_field = {}
missing_occupations = []
for code in code_to_occupation:
    occupation = code_to_occupation[code]
    
    if occupation in occupation_to_field.keys():
        code_to_field[code] = occupation_to_field[occupation]
    else:
        missing_occupations.append(occupation)
```

```python
# save dictionary in json format
with open('./data/dict_wikicode_to_field.json', 'w') as fp:
    json.dump(code_to_field, fp)
```
